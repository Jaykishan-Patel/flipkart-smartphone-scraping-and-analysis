Smartphone Dataset Summary
## General Overview
Rows: 984 (smartphone models & variants)
Columns: 10
Coverage: Budget (₹4,399) to Flagships (₹1,57,900+)
Brands Included: Apple, Samsung, OnePlus, Motorola, vivo, realme, OPPO, POCO, Infinix, iQOO, Lava, Tecno, itel, CMF by Nothing, etc.
Key Trend: Many duplicate models across different colors and storage variants.

##  Column-wise Observations
# model_name
Brand + model + variant details.
Observation: Multiple duplicate models across storage/colors. Useful for brand presence & variant pricing analysis.
# price
Smartphone price in ₹ (string format).
Observation: Wide range (₹4,399 – ₹1,57,900+). Some missing/inconsistent entries. Good for budget vs flagship segmentation.
# avg_rating
Average customer rating (float).
Observation: Mostly between 4.1–4.6 → overall positive feedback. Can compare ratings by brand/segment.
# reviews
Text like “47,939 Ratings & 3,315 Reviews”.
Observation: Needs parsing into ratings count + reviews count. Review counts vary widely, showing popularity vs new launches.
# memory
RAM + ROM + expandable storage.
Observation: Semi-structured string. Must split into RAM, internal storage, expandable storage for analysis of memory vs price.
# display
Screen size + resolution.
Observation: Mostly 6.5–6.9 inches, with “HD+” vs “Full HD+”. Can be used for screen trends across price segments.
# battary (⚠ actually camera details
Example: “50MP Rear Camera | 8MP Front Camera”.
Observation: Rear usually 50MP; higher MP setups in premium phones. Needs parsing into rear MP + front MP.
# processor (⚠ actually battery capacity)
Example: “5100 mAh Battery”.
Observation: Most phones have 5000–7000 mAh. Good for battery size vs. price/rating analysis.
# upto_exchange
Exchange/discount offer value.
Observation: Ranges from ~₹4,000 to ₹18,000+. Can study discount strategy per brand/segment.
