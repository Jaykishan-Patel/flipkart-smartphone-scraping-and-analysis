```python
with open('all_pages.html','r',encoding='utf-8') as f:
    html=f.read()
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
containers=soup.find_all('div',{'yKfJKb row'})
```

```python
name=[]
price=[]
avg_rating=[]
votes_for_rating=[]
memory=[]
display=[]
camera=[]
battery=[]
processor=[]
upto_exchange=[]
for i in containers:
    try:
        name.append(i.find('div',{'class':'KzDlHZ'}).text)
    except:
        name.append(np.nan)
    try:
        price.append(i.find('div',{'class':'Nx9bqj _4b5DiR'}).text)
    except:
        price.append(np.nan)
    try:
        avg_rating.append(i.find('div',{'class':'XQDdHH'}).text)
    except:
        avg_rating.append(np.nan)  
    try:
        votes_for_rating.append(i.find('span',{'class':'Wphh3N'}).text)
    except:
        votes_for_rating.append(np.nan)
    try:
        upto_exchange.append(i.find('div',{'class':'M4DNwV'}).text)
    except:
        upto_exchange.append(np.nan)
    x=i.find_all('li',{'class':'J+igdf'})
    try:
        memory.append(x[0].text)
    except:
        memory.append(np.nan)
    try:
        display.append(x[1].text)
    except:
        display.append(np.nan)
    try:
        camera.append(x[2].text)
    except:
        camera.append(np.nan)
    try:
        battery.append(x[3].text)
    except:
        battery.append(np.nan)
    try:
        processor.append(x[4].text)
    except:
        processor.append(np.nan)
```

```python
df=pd.DataFrame({
    "model_name":name,
    "price":price,
    "avg_rating":avg_rating,
    "reviews":votes_for_rating,
    "memory":memory,
    "display":display,
    "battary":battery,
    "processor":processor,
    "upto_exchange":upto_exchange
})
```
